### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python

csvFile['horsepower'] = pd.to_numeric(csvFile['horsepower'], errors='coerce')
csvFile['hp_to_weight'] = csvFile['horsepower'] / csvFile['weight']

median_mpg = csvFile['mpg'].median()

above_median_cars = csvFile[csvFile['mpg'] > median_mpg]

max_ratio = above_median_cars['hp_to_weight'].max()
best_cars = above_median_cars[above_median_cars['hp_to_weight'] == max_ratio]

print(f"Median MPG: {median_mpg:.1f}")
print(f"\nCar(s) with best hp/weight ratio ({max_ratio:.4f}) among above-median MPG cars:")
print(best_cars[['name', 'mpg', 'horsepower', 'weight', 'hp_to_weight']].to_string(index=False))
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('/Cars.csv')
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')

avg_mpg = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")

ax = sns.lineplot(
    data=avg_mpg,
    x='model_year',
    y='mpg',
    hue='origin',
    style='origin',
    markers=['o', 's', 'D'],
    markersize=10,
    linewidth=2.5,
    palette='bright',
    dashes=False
)

plt.title('Evolution of Average MPG by Origin (1970-1982)', fontsize=14, pad=20)
plt.xlabel('Model Year', fontsize=12)
plt.ylabel('Average MPG', fontsize=12)
plt.xticks(sorted(df['model_year'].unique()))

plt.legend(
    title='Origin',
    bbox_to_anchor=(1.05, 1),
    loc='upper left',
    frameon=True
)

for origin in avg_mpg['origin'].unique():
    last_year_data = avg_mpg[(avg_mpg['model_year'] == 1982) & (avg_mpg['origin'] == origin)]
    if not last_year_data.empty:
        ax.text(
            x=1982.2,
            y=last_year_data['mpg'].values[0],
            s=f"{last_year_data['mpg'].values[0]:.1f}",
            va='center',
            fontsize=10
        )

plt.tight_layout()
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('/Cars.csv')
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')

plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(
    data=df,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    hue_order=['japan', 'europe', 'usa'],
    palette='bright',
    alpha=0.8
)

plt.title('Vehicle Weight vs. Horsepower by Origin (Size = MPG)', fontsize=14)
plt.xlabel('Horsepower', fontsize=12)
plt.ylabel('Weight (lbs)', fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
import pandas as pd

df = pd.read_csv('/Cars.csv')

model_counts = df['name'].value_counts()
multi_year_models = model_counts[model_counts > 1].index

consistency = df[df['name'].isin(multi_year_models)].groupby('name')['mpg'].agg(['std', 'count', 'mean'])

consistent_cars = consistency[(consistency['std'] < 1.0) & (consistency['count'] >= 2)]

if not consistent_cars.empty:
    min_std = consistent_cars['std'].min()
    most_consistent = consistent_cars[consistent_cars['std'] == min_std]

    # Bonus: Sort results
    result = consistent_cars.sort_values(['count', 'mean'], ascending=[False, False])
    result.columns = ['MPG_StdDev', 'Appearances', 'Avg_MPG']

    print("Most consistent car models (std < 1.0):")
    print(result[['Appearances', 'Avg_MPG', 'MPG_StdDev']].to_string())
else:
    print("No models meet the consistency criteria")
```
